```python
#Loading MNIST dataset  containing training samples of 60,000 and testing sample of 10,000
#Tensorflow containg MNIST dataset which can be loaded using Keras
mnist = tf.keras.datasets.mnist
mnist
```

```python
#Load the data and split it into train and test sets
(X_train,y_train),(X_test,y_test) = mnist.load_data()
```

```python
#Normalizing the data || pre-processing Step as the images are in grey scale (1 Channel --> 0 to 255), not coloured image
#In grey scale pixel value close to 0 denote the pixel is close to black in colour and reverse is true for white colour
X_train = tf.keras.utils.normalize(X_train, axis=1)
X_test = tf.keras.utils.normalize(X_test, axis=1)

print("Index 0th val from train set : " ,y_train[0])
plt.imshow(X_train[0], cmap = plt.cm.binary) 
plt.show()

print("Index 0th val from test set : " ,y_test[0])
plt.imshow(X_test[0], cmap = plt.cm.binary) 
plt.show()
```

**Resizing img to make it suitable for applying Convolution operation**

```python
import numpy as np
X_trainr = np.array(X_train).reshape(-1,28,28,1) #increasing size for one dimension kernel operation
X_testr = np.array(X_test).reshape(-1,28,28,1)
print("Training sample dimension:",X_trainr.shape)
print("Testing sample dimension:",X_testr.shape)
```

**Creating Deep Nural Network**

#Training on 60,000 sample data of MNIST dataset

```python
!pip install tensorflow==1.14 
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, Activation, Flatten, Conv2D, MaxPooling2D
```

```python
X_trainr.shape[1:]
```

```python
#creating model
model = Sequential()

#First convolution layer
model.add(Conv2D(64,(3,3), input_shape = (28, 28, 1)))
model.add(Activation("relu"))
model.add(MaxPooling2D(pool_size = (2,2)))

#2nd Convolution layer
model.add(Conv2D(64,(3,3)))
model.add(Activation("relu"))
model.add(MaxPooling2D(pool_size = (2,2)))

#3rd Convolution layer
model.add(Conv2D(64,(3,3)))
model.add(Activation("relu"))
model.add(MaxPooling2D(pool_size = (2,2)))

#Fully connected layer "1"
model.add(Flatten())
model.add(Dense(64))
model.add(Activation("relu"))

#Fully connected layer "2"
model.add(Dense(32))
model.add(Activation("relu"))

#Last Fully connected layer
model.add(Dense(10))
model.add(Activation("softmax"))
```

```python
model.summary()
```

```python
model.compile(loss = "sparse_categorical_crossentropy", optimizer="adam" , metrics=["accuracy"])
```

```python
model.fit(X_trainr, y_train, epochs=5, validation_split=0.3)
```

```python
model.save("/content/drive/MyDrive/Blu  Internship/default_model.hdf5")
```

```python
#Evaluating on Test dataset of MNIST
test_loss, test_acc = model.evaluate(X_testr,y_test)
print("Test loss on 10,000 test sample : ", test_loss)
print("Validation accuracy on 10,000 test sample : ", test_acc)
```

```python
predictions = model.predict(X_testr)
predictions
```

```python
#Print our predictions as number labels
print("our predictions as number labels : ")
print(np.argmax(predictions, axis=1))
#Print the actual labels
print("the actual labels : ")
print(y_test)
#let's check with image
plt.imshow(X_test[0])
```
